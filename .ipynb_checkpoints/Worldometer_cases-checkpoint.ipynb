Fitting the data manually:

```python
def gaussian_func(x, a, x0, sigma,c):
    return a * np.exp(-(x-x0)**2/(2*sigma**2)) + c
```

```python
def predict_case_total(df):
    #Make the labels for x-axis
    xlabels = [dates[20*i] for i in range(round(len(dates)/20))]
    xlabels += ['May 5', 'May 25', 'June 14']
    xlabels.insert(0, '')
    #Daily new cases plot, grab data from df
    x = np.array([i + 1 for i in range(len(df['Date']))])
    y = np.array([d for d in df['New Cases']])
    #Make arrays to plug into the gaussian fit
    x_predict = np.linspace(1,120,1200)
    x_predict_day = np.linspace(1,120,120)
    #initial guess for gaussian parameters
    initial_guess = [1,20,75,0]
    #fit gaussian
    popt, pcov = curve_fit(gaussian_func, x, y,p0=initial_guess)
    #set up plot
    sns.set_style('whitegrid')
    f, ax = plt.subplots(1,1)
    plt.xlabel('Date')
    plt.ylabel('Daily New Cases')
    plt.title('Daily New Cases of COVID-19')
    ax.set_xticklabels(xlabels)
    #plot the real daily case data as scatter plot
    sns.scatterplot(x, y, data = df, label = 'Real New Cases')
    #plot the predicted daily cases as a line
    sns.lineplot(x_predict, abs(gaussian_func(x_predict, *popt)), data = df, label = 'Predicted New Cases', color = 'red', alpha = 0.5)
    plt.show()
    
    #total cases is just the sum of daily cases, create that list
    case_list = [abs(gaussian_func(x_predict_day[i], *popt)) for i in range(120)]
    case_list_sum = [sum(case_list[:i]) for i in range(120)]
    #set up second plot
    f, ax = plt.subplots(1,1)
    plt.xlabel('Date')
    plt.title('Total Cases of COVID-19')
    ax.set_xticklabels(xlabels)
    #plot real total cases as scatter plot
    sns.scatterplot(x, df['Total Cases'], data=df, label = 'Real Total Cases')
    #plot predicted total cases as a line
    sns.lineplot(x_predict_day, case_list_sum, data = df, color = 'red', alpha = 0.5, label = 'Predicted Total Cases')
    plt.show()
    #print(int(max(case_list_sum)))
predict_case_total(df)
```

```python
def predict_death_total(df):
    #Make the labels for x-axis
    xlabels = [dates[20*i] for i in range(round(len(dates)/20))]
    xlabels += ['May 5', 'May 25', 'June 14']
    xlabels.insert(0, '')
    #Daily new deaths plot, grab data from df
    x = np.array([i + 1 for i in range(len(df['Date']))])
    y = np.array([d for d in df['New Deaths']])
    #Make arrays to plug into the gaussian fit
    x_predict = np.linspace(1,120,1200)
    x_predict_day = np.linspace(1,120,120)
    #initial guess for gaussian parameters
    initial_guess = [1,20,75,0]
    #fit gaussian
    popt, pcov = curve_fit(gaussian_func, x, y,p0=initial_guess)
    #set up plot
    sns.set_style('whitegrid')
    f, ax = plt.subplots(1,1)
    plt.xlabel('Date')
    plt.ylabel('Daily New Deaths')
    plt.title('Daily New Deaths from COVID-19')
    ax.set_xticklabels(xlabels)
    #plot the real daily death data as scatter plot
    sns.scatterplot(x, y, data = df, label = 'Real New Deaths')
    #plot the predicted daily deaths as a line
    sns.lineplot(x_predict, abs(gaussian_func(x_predict, *popt)), data = df, label = 'Predicted New Deaths', color = 'red', alpha = 0.5)
    plt.show()
    
    #total deaths is just the sum of daily deaths, create that list
    death_list = [abs(gaussian_func(x_predict_day[i], *popt)) for i in range(120)]
    death_list_sum = [sum(death_list[:i]) for i in range(120)]
    #set up second plot
    f, ax = plt.subplots(1,1)
    plt.xlabel('Date')
    plt.title('Total Deaths from COVID-19')
    ax.set_xticklabels(xlabels)
    #plot real total deaths as scatter plot
    sns.scatterplot(x, df['Total Deaths'], data=df, label = 'Real Total Deaths')
    #plot predicted total deaths as a line
    sns.lineplot(x_predict_day, death_list_sum, data = df, color = 'red', alpha = 0.5, label = 'Predicted Total Deaths')
    plt.show()
    print('Predicted total deaths on June 14: ')
    print(int(death_list_sum[-1]))
predict_death_total(df)
```

Fitting data using a Gaussian Regression Process machine learning model:

```python
def machine_cases(df):
    #First do new cases
    xlabels = [dates[20*i] for i in range(round(len(dates)/20))]
    xlabels += ['May 5', 'May 25', 'June 14']
    xlabels.insert(0, '')
    #make the arrays for fitting and plotting since the model requires 1-D data
    X = np.array([i+1 for i in range(len(df['Date']))]).reshape(-1, 1)
    y = np.array(df[['New Cases']]).reshape(-1, 1)
    X_long = np.array([i+1 for i in range(120)]).reshape(-1, 1)
    plt_X_long = [i+1 for i in range(120)]
    plt_X = [i+1 for i in range(len(df['Date']))]
    #create and fit the GPR Model
    kernel = Matern() + ExpSineSquared()
    gpr = GaussianProcessRegressor(kernel = kernel, random_state=1).fit(X, y)
    #predict the daily cases
    prediction_vals = [val[0] for val in gpr.predict(X_long)]
    #plot the data
    sns.set_style('whitegrid')
    f, ax = plt.subplots(1,1)
    sns.lineplot(plt_X_long, prediction_vals, data=df, label = 'GPR Predicted New Cases', color = 'red')
    sns.scatterplot(plt_X, df['New Cases'], data=df, label = 'Actual New Cases')
    plt.title('Daily New Cases of COVID-19')
    plt.xlabel('Date')
    ax.set_xticklabels(xlabels)
    plt.show()
    #Now do total cases as a sum of new cases
    prediction_total = [sum(prediction_vals[:i]) for i in range(len(prediction_vals))]
    actual_total = df['Total Cases']
    f, ax = plt.subplots(1,1)
    sns.lineplot(plt_X_long, prediction_total, data=df, label = 'GPR Predicted Total Cases', color = 'red')
    sns.scatterplot(plt_X, actual_total, data = df, label = 'Actual Total Cases')
    plt.title('Total Cases of COVID-19')
    plt.xlabel('Date')
    ax.set_xticklabels(xlabels)
    plt.show()
    
    

machine_cases(df)
```

We can see the prediction is more concave past the peak than we would expect from a gaussian.

```python
def machine_deaths(df):
    #First do new deaths
    xlabels = [dates[20*i] for i in range(round(len(dates)/20))]
    xlabels += ['May 5', 'May 25', 'June 14']
    xlabels.insert(0, '')
    #make the arrays for fitting and plotting since the model requires 1-D data
    X = np.array([i+1 for i in range(len(df['Date']))]).reshape(-1, 1)
    y = np.array(df[['New Deaths']]).reshape(-1, 1)
    X_long = np.array([i+1 for i in range(120)]).reshape(-1, 1)
    plt_X_long = [i+1 for i in range(120)]
    plt_X = [i+1 for i in range(len(df['Date']))]
    #create and fit the GPR Model
    kernel = Matern() + ExpSineSquared() 
    gpr = GaussianProcessRegressor(kernel = kernel, random_state=1).fit(X, y)
    #predict the daily cases
    prediction_vals = [val[0] for val in gpr.predict(X_long)]
    #plot the data
    sns.set_style('whitegrid')
    f, ax = plt.subplots(1,1)
    sns.lineplot(plt_X_long, prediction_vals, data=df, label = 'GPR Predicted New Deaths', color = 'red', alpha = 0.5)
    sns.scatterplot(plt_X, df['New Deaths'], data=df, label = 'Actual New Deaths')
    plt.title('Daily New Deaths from COVID-19')
    plt.xlabel('Date')
    ax.set_xticklabels(xlabels)
    plt.show()
    #Now do total cases as a sum of new cases
    prediction_total = [sum(prediction_vals[:i]) for i in range(len(prediction_vals))]
    actual_total = df['Total Deaths']
    f, ax = plt.subplots(1,1)
    sns.lineplot(plt_X_long, prediction_total, data=df, label = 'GPR Predicted Total Deaths', color = 'red', alpha = 0.5)
    sns.scatterplot(plt_X, actual_total, data = df, label = 'Actual Total Deaths')
    plt.title('Total Deaths from COVID-19')
    plt.xlabel('Date')
    ax.set_xticklabels(xlabels)
    plt.show()
    print('Predicted total deaths on June 14: ')
    print(int(prediction_total[-1]))
    
machine_deaths(df)
    
```

The deaths are pretty similar to the cases, just far smaller in number. The machine learning model does predict a slightly higher amount of total deaths than the manually fitted model. This is most likely because the data does not fit to a perfect gaussian. Also the machine learning model is only fit with around 65 data points so it may not have a perfect idea of the data.
